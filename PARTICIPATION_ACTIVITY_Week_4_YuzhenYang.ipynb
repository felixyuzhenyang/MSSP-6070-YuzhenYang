2. Complete the following exercise:


a. Find the 95th percentile of earthquake magnitude in Japan using magType of 'mb'



```python
parsed_data = pd.read_csv(f"{folder_path}/parsed.csv")
# print(parsed_data.head())
# print(parsed_data.info())
# print(parsed_data.describe())

japan_mb_quakes = parsed_data[
    (parsed_data["parsed_place"] == "Japan") &
    (parsed_data["magType"] == "mb")
]

per_95 = japan_mb_quakes["mag"].quantile(0.95)

print(f"The 95% percentile of earthquake magnitude in Japan is: {per_95:.2f}")
```

b. Find the percentage of earthquakes in Indonesia that were coupled with tsunamis.

```python
id_quakes = parsed_data[parsed_data["parsed_place"] == "Indonesia"]
tsunami_coupled_quakes = id_quakes["tsunami"].sum()

per_tsunami = (tsunami_coupled_quakes / len(id_quakes)) * 100

print(f"The percentage of earthquakes in Indonesia that were coupled with tsunamis is: {per_tsunami:.2f}%")
```
