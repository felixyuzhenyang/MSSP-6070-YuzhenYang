##Lambda Functions:

This notebook is an example of how lamda functions works.  For more information see https://docs.python.org/3/reference/expressions.html#lambda'

As we dicussed in class, Lambda Functions which are anonymous functions meaning they are not bound by names. The unamed object behaves like a function object.   

The format of a Lambda function is: **lambda ::= [parameter lists] ":" expression**.  Note that functions created with lambda expressions cannot contain statements or annotations; lambda functions contain a single valid python expression.  Also, the Lambda function can only have one expression.

**Definition: Expression:** A piece of syntax which can be evaluated to some value. In other words, an expression is an accumulation of expression elements like literals, names, attribute access, operators or function calls which all return a value. In contrast to many other languages, not all language constructs are expressions. There are also statements which cannot be used as expressions, such as while. Assignments are also statements, not expressions.

See: https://docs.python.org/3/glossary.html#term-expression

Lamda functions are shorthand representations of functions that allow us to write concise code without multiple lines.

```python
import pandas as pd
```

```python
df=pd.read_csv('Student_Performance_Data.csv') #Input the Student Performance data set into a Pandas Data Frame.
df
```

```python
#The lambda declaration creates an non-named object that acts as a function in that it is given values, evaluates those
# values, and returns a result.

x = lambda a,b : a +b+ 10   #defines and object with one or more parameters, with one expression that returns a result.

print(x(15,10))
```

```python
#We can also return lambda functions as a result
def find_number (x): #Define a function
    return lambda x: x+(2**3) #return lambda function and assign it to a variable

new_number_one = find_number(0) #new_number_one calls find_number function and is assigned the lambda function lambda x: x+(2**3)
new_number_two = find_number(0) #new_number_two calls find_number function and is assigned the lambda function lambda x: x+(2**3)
print(new_number_one(3), new_number_two(2)) #Send the newly defined function parameters and evaluate the results. 3+(2**3) = 11, 2+(2**3) = 10.
```

```python
#Lamda functions can be used inside statements to evaluate and expression requring results.
df["reading score"] = df["reading score"].apply(lambda x: x-2) #Changes the reading score by subtracting 2 from the current values in the series.

print(df["reading score"])

```
